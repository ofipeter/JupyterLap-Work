```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Question 1: Day 2

# Creating a Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [50, 30, 40, 60]
plt.bar(categories, values, color= 'gold')
plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Question 2: Day 2

# Creating a Scatter Plot
import seaborn as sns
import matplotlib.pyplot as plt

x = [5, 10, 15, 20, 25]
y = [10, 15, 20, 25, 30]

sns.scatterplot(x=[5,10,15,20,25], y= [10,15,20,25,30])
plt.title("Scatter Plot for X and Y")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# Question 3: Day 2

# Creating subplots
fig,axs = plt.subplots(1,2, figsize=(10,5))

# Visitors
days = [1, 2, 3, 4, 5]
visitors = [100, 150, 200, 150, 300]
axs[0].plot(days, visitors)
axs[0].set_title('Visitors Over 5 Days')

# Sales
sales = [20, 28, 35, 27, 40]
axs[1].bar(days, sales)
axs[1].set_title('Sales Over 5 Days')

plt.tight_layout()
plt.show()
```

```python
# Question 4: Day 2

# Creating Grouped Bar Chart
quarters = ["Q1", "Q2", "Q3", "Q4"]
profits_2022 = [10000, 12000, 11000, 13000]
profits_2023 = [15000, 14000, 16000, 17000]

X_qrts = np.arange(len(quarters))

plt.bar(X_qrts-0.2, profits_2022, 0.4, label= 2022)
plt.bar(X_qrts+0.2, profits_2023, 0.4, label= 2023)
plt.xticks(X_qrts, quarters)
plt.xlabel("Quarters")
plt.ylabel("Profits")
plt.title("Profit in Quarters For 2022 and 2023")
plt.legend()
plt.show()
```
