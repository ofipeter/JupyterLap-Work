```python
# Question 1: Day 1

import matplotlib.pyplot as plt
```

```python
# Creating a Simple Line Plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# Question 2 & 3: Day 1

# Change the color of the line in the plot from question 1 to 'green' and make the line dashed
plt.plot(x, y, linestyle= 'dashed', color= 'g')
plt.show()
```

```python
# Question 4: Day 1

# Real-World Scenario
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [200, 300, 250, 450, 500]
plt.plot(months, sales, color= 'red')
plt.title("Small Business Monthly Sales")
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```
